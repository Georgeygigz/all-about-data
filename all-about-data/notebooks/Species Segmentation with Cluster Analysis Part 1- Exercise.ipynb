## Import the relevant libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
sns.set()
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
iris_dataset_data = pd.read_csv(os.path.join(os.path.pardir,'data','raw','iris-dataset.csv'))
iris_dataset_data.head()
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
plt.scatter(iris_dataset_data['sepal_length'],iris_dataset_data['sepal_width'])
plt.xlabel('Sepal length',fontsize=20)
plt.ylabel('Sepal width',fontsize=20);
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x = iris_dataset_data.copy()
x
```

```python
kmeans = KMeans(2)
```

```python
kmeans.fit(x)
```

```python
iris_dataset_data_clustured = iris_dataset_data.copy()
```

```python
iris_dataset_data_clustured['Cluster'] = kmeans.fit_predict(x)
```

```python
plt.scatter(iris_dataset_data_clustured['sepal_length'],iris_dataset_data_clustured['sepal_width'],c=iris_dataset_data_clustured['Cluster'],cmap='rainbow');
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

# Clustering (scaled data)

```python
kmeans_scaled = KMeans(2)
```

```python
kmeans_scaled.fit(x_scaled)
```

```python
iris_dataset_data_clustured = iris_dataset_data.copy()
```

```python
iris_dataset_data_clustured['Cluster'] = kmeans_scaled.fit_predict(x_scaled)
```

```python
plt.scatter(iris_dataset_data_clustured['sepal_length'],iris_dataset_data_clustured['sepal_width'],c=iris_dataset_data_clustured['Cluster'],cmap='rainbow');
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []
c_nums = 10
for i in range(1,c_nums):
    kmeans = KMeans(i)
    kmeans.fit_predict(x_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

### The Elbow Method

```python
plt.plot(range(1,10),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```
